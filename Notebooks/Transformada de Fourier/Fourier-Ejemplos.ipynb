### Ejemplos de series de Fourier y transformadas de Fourier

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

#### Series de Fourier:
La serie de Fourier de la funci칩n signo 
$$
 f(x) := \left\{ \begin{array}{cc}
     -1,& - \pi \leq x < 0  \\
     1,&   0 \leq x \leq \pi
\end{array} \right.,  
$$
est치 dada por
$$
 f(x) = \sum_{k=1}^{\infty} \frac{4}{\pi} \frac{\sin[(2k-1)x]}{(2k-1)}.   
$$

```python
#Definici칩n de la serie de Fourier:
def Fourier(n,x):
    if (n==1):
        fourier = (4/(np.pi))*np.sin(x) 
    elif (n>1):
        fourier = (4/(np.pi))*np.sin(x) 
        for j in range(2,n+1,1):
            fourier = fourier + (4/(np.pi))*(np.sin((2*j-1)*x)/(2*j-1)) 
    return fourier

#Partici칩n del dominio x:
x1 = np.linspace(-3*np.pi,3*np.pi,1000) 
x21 = np.linspace(-np.pi,0,100) 
x22 = np.linspace(0,np.pi,100) 
```

```python
#Gr치fica de la funci칩n y su serie de Fourier:
plt.figure(figsize = (11,4))
plt.plot(x21,-1*np.ones(len(x21)),color = 'black',label = '$f(x)$',linewidth = 3)
plt.plot(x22,np.ones(len(x22)),color = 'black',linewidth = 3)
plt.plot(x1,Fourier(1,x1), label = '$s_1(x)$')
plt.plot(x1,Fourier(2,x1), label = '$s_2(x)$', linestyle= '--')
plt.plot(x1,Fourier(3,x1), label = '$s_3(x)$', linestyle= '-.')
plt.plot(x1,Fourier(4,x1), label = '$s_4(x)$', linestyle= '-.')
plt.xlabel('$x$',fontsize = 15)
plt.ylabel('$s_n(x)$',fontsize = 15)
plt.legend(loc = 'best',fontsize = 12)
plt.grid(True, linestyle = '--')
plt.xticks(ticks = [-3*np.pi,-2*np.pi,-np.pi, 0,np.pi,2*np.pi,3*np.pi],labels = ['$-3\\pi$','$-2\\pi$', '$-\\pi$','0','$\\pi$', '$2\\pi$','$3\\pi$' ], fontsize = 12)
plt.yticks(ticks = [-1,1], labels = [-1,1],  fontsize = 10)
plt.savefig('EjemploFourier2.pdf',bbox_inches = "tight")
```

#### Transformada de Fourier:
La transformada de Fourier del pulso cuadrado
$$
f(x) = \left\{ \begin{array}{cl}
     1,& |x|<a  \\
     0,& |x|> a
\end{array} \right.,  
$$
 est치 dada por
$$
\hat{f}(\omega) = \sqrt{\frac{2}{\pi}} \frac{\sin(\omega a)}{\omega}.
$$

```python
#Transformada de Fourier 1:
a = 5
k = np.linspace(-10,10,1000)
t1 = np.linspace(-5,5,100)
t2 = np.linspace(-10,-5,100)
t3 = np.linspace(5,10,100)

TFourier1 = np.sqrt(2/np.pi)*(np.sin(k*a))/k
```

```python
#Gr치fica de la funci칩n y su transformada de Fourier:
plt.figure(figsize = (9,4))
plt.subplot(1,2,1)
plt.plot(t1,np.ones(len(t1)),color = 'red',linewidth = 2)
plt.plot(t2,np.zeros(len(t2)),color = 'red',linewidth = 2)
plt.plot(t3,np.zeros(len(t3)),color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$x$', fontsize = 15)
plt.ylabel('$f\,(x)$', fontsize = 15)
plt.subplot(1,2,2)
plt.plot(k,TFourier1, color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$\\omega$', fontsize = 15)
#plt.ylim(-0.1,2)
plt.ylabel('$\\hat{f}\,(\\omega)$', fontsize = 15, labelpad = -5)
plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.26)
plt.savefig('EjemploTransformada1.pdf',transparent = True,bbox_inches = "tight")
```

La transformada de Fourier de la gaussiana
$$
f(x) = n e^{-\alpha x^2}, \quad  \alpha > 0,
$$
est치 dada por
$$
\hat{f}(\omega) = \frac{n}{\sqrt{2\alpha}} e^{- \left( \frac{\omega^2}{4\alpha} \right)}.   
$$

```python
#Transformadas de Fourier 2:
a1 = 1
a2 = 0.25
n = 1
k = np.linspace(-10,10,1000)

f1 = n*(np.exp(-a1*(k**2)))
f2 = n*(np.exp(-a2*(k**2)))

TFourier21 = (n/np.sqrt(2*a1))*np.exp((-k**2)/(4*a1))
TFourier22 = (n/np.sqrt(2*a2))*np.exp((-k**2)/(4*a2))
```

```python
#Gr치fica de la funci칩n y su transformada de Fourier:
plt.figure(figsize = (9,4))
plt.subplot(1,2,1)
plt.plot(k,f1,color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$x$', fontsize = 15)
plt.ylabel('$f\,(x)$', fontsize = 15)
plt.subplot(1,2,2)
plt.plot(k,TFourier21, color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$\\omega$', fontsize = 15)
plt.ylabel('$\\hat{f}\,(\\omega)$', fontsize = 15)
plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.26)
plt.savefig('EjemploTransformada2.pdf', transparent = True,bbox_inches = "tight")
```

```python
#Gr치fica de la funci칩n y su transformada de Fourier:
plt.subplot(1,2,1)
plt.plot(k,f2,color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$x$', fontsize = 15)
plt.ylabel('$f\,(x)$', fontsize = 15)
plt.subplot(1,2,2)
plt.plot(k,TFourier22, color = 'red',linewidth = 2)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.xlabel('$\\omega$', fontsize = 15)
plt.ylabel('$\\hat{f}\,(\\omega)$', fontsize = 15)
plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.26)
plt.savefig('EjemploTransformada3.pdf', transparent = True,bbox_inches = "tight")
```
